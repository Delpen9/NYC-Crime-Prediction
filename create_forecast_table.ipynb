## Get all unique combinations of column values and assign them a complaint count of 0

#### np.repeat and np.tile will be used to generate this dataframe

#### We want all complaint possibilities to be accounted for. We will need to manage the class imbalance (1:7) in the modeling.

```python
total_count = 1

num_unique_year = 1
num_unique_month = 1
num_unique_hour = 1
num_unique_sex = 1
num_unique_nta = 1
num_unique_weekday = 1
num_unique_age = 1
num_unique_is_black = 1

if 'year' in groupby_columns:
    num_unique_year = df_crime_complaints_dataset_1['year'].nunique()
    total_count *= num_unique_year
    
if 'month' in groupby_columns:
    num_unique_month = df_crime_complaints_dataset_1['month'].nunique()
    total_count *= num_unique_month
    
if 'hour' in groupby_columns:
    num_unique_hour = df_crime_complaints_dataset_1['hour'].nunique()
    total_count *= num_unique_hour

if 'victim_sex' in groupby_columns:
    num_unique_sex = df_crime_complaints_dataset_1['victim_sex'].nunique()
    total_count *= num_unique_sex

if 'NTA' in groupby_columns:
    num_unique_nta = df_crime_complaints_dataset_1['NTA'].nunique()
    total_count *= num_unique_nta
    
## Extras
if 'weekday' in groupby_columns:
    num_unique_weekday = df_crime_complaints_dataset_1['weekday'].nunique()
    total_count *= num_unique_weekday
    
if 'victim_age_range' in groupby_columns:
    num_unique_age = df_crime_complaints_dataset_1['victim_age_range'].nunique()
    total_count *= num_unique_age
    
if 'is_black' in groupby_columns:
    num_unique_is_black = df_crime_complaints_dataset_1['is_black'].nunique()
    total_count *= num_unique_is_black
```

```python
all_combinations_columns = []

if 'year' in groupby_columns:
    # year
    unique_year_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['year'].unique()),
        int(total_count / (num_unique_nta * num_unique_sex * num_unique_hour * num_unique_month * num_unique_year))
    )
    unique_year_column = np.repeat(
        unique_year_column,
        int(num_unique_nta * num_unique_sex * num_unique_hour * num_unique_month)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_year_column)
    
if 'month' in groupby_columns:
    # month
    unique_month_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['month'].unique()),
        int(total_count / (num_unique_nta * num_unique_sex * num_unique_hour * num_unique_month))
    )
    unique_month_column = np.repeat(
        unique_month_column,
        int(num_unique_nta * num_unique_sex * num_unique_hour)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_month_column)
    
if 'weekday' in groupby_columns:
    # weekday
    unique_weekday_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['weekday'].unique()),
        int(total_count / (num_unique_nta * num_unique_sex *\
                           num_unique_hour * num_unique_month * num_unique_year * num_unique_weekday))
    )
    unique_weekday_column = np.repeat(
        unique_weekday_column,
        int(num_unique_nta * num_unique_sex * num_unique_hour * num_unique_month * num_unique_year)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_weekday_column)
    
if 'hour' in groupby_columns:
    # hour
    unique_hour_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['hour'].unique()),
        int(total_count / (num_unique_nta * num_unique_sex * num_unique_hour))
    )
    unique_hour_column = np.repeat(
        unique_hour_column,
        int(num_unique_nta * num_unique_sex)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_hour_column)
    
if 'victim_age_range' in groupby_columns:
    # victim age
    unique_age_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['victim_age_range'].unique()),
        int(total_count / (num_unique_nta * num_unique_sex *\
                           num_unique_hour * num_unique_month * num_unique_year * num_unique_weekday * num_unique_age))
    )
    unique_age_column = np.repeat(
        unique_age_column,
        int(num_unique_nta * num_unique_sex * num_unique_hour * num_unique_month * num_unique_year * num_unique_weekday)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_age_column)
    
if 'is_black' in groupby_columns:
    # is black?
    unique_is_black_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['is_black'].unique()),
        int(total_count / (num_unique_nta * num_unique_sex *\
                           num_unique_hour * num_unique_month * num_unique_year * num_unique_weekday * num_unique_age * num_unique_is_black))
    )
    unique_is_black_column = np.repeat(
        unique_is_black_column,
        int(num_unique_nta * num_unique_sex * num_unique_hour *\
            num_unique_month * num_unique_year * num_unique_weekday * num_unique_age)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_is_black_column)
    
if 'victim_sex' in groupby_columns:
    # victim_sex
    unique_sex_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['victim_sex'].unique()),
        int(total_count / (num_unique_nta * num_unique_sex))
    )
    unique_sex_column = np.repeat(
        unique_sex_column,
        int(num_unique_nta)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_sex_column)
    
if 'NTA' in groupby_columns:
    # NTA
    unique_nta_column = np.tile(
        np.sort(df_crime_complaints_dataset_1['NTA'].unique()),
        int(total_count / num_unique_nta)
    ).reshape(-1, 1)
    
    all_combinations_columns.append(unique_nta_column)
```
